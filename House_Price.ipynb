```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("data.csv")
```

```python
data.head()
```

```python
temp_data=data.drop(['date','street'],axis=1)
temp_data.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
temp_data['city'] = encoder.fit_transform(temp_data['city'])
temp_data['statezip'] = encoder.fit_transform(temp_data['statezip'])
temp_data['country'] = encoder.fit_transform(temp_data['country'])
temp_data.head()

```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
imputer.fit(temp_data)
```

```python
temp_data.info()
```

```python
# Train_Test_Split
from sklearn.model_selection import train_test_split
train_data,test_data=train_test_split(temp_data,test_size=0.2,random_state=42)
train_data_features=train_data.drop("price",axis=1)
train_data_labels=train_data['price'].copy()
test_data_labels=test_data['price'].copy()
```

```python
# Model Selection
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error  
import numpy as np  

model = LinearRegression()
model.fit(train_data_features, train_data_labels)


predictions_result = model.predict(train_data_features)  
predictions_result
mse = mean_squared_error(train_data_labels, predictions_result)  # Corrected to use train_data_labels
print(mse)
```

```python
from sklearn.metrics import r2_score
accuracy2=r2_score(predictions_result,train_data_labels)
print(accuracy2*100)
```
